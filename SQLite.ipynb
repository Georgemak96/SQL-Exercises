```python
import sqlite3 as sq3
import pandas.io.sql as pds
con = sq3.connect("Products_Components.db")

with con:
    con.execute("CREATE TABLE Products (ProductID INT PRIMARY KEY NOT NULL, ProductName CHAR,Cost REAL, Price REAL)")
    
    data_products = [(123, "Small Monkey", 2.50, 5.95),
                     (156, "Pink Kitten", 3.10, 6.00)]
    for i in data_products:
        con.execute("INSERT INTO Products VALUES (?,?,?,?)", i)

    
    con.execute("CREATE TABLE Components (ProductID INT NOT NULL, CompID CHAR NOT NULL, CompQty REAL,"
                "PRIMARY KEY (ProductID,CompID),"
                "FOREIGN KEY (ProductID) REFERENCES Products(ProductID))")
    
    data_Components = [(123, "ST01", 30),
                       (123, "G56", 2),
                       (123, "FF77", 0.3),
                       (156, "ST01", 45),
                       (156, "G120", 2),
                       (156, "FF88", 0.35),
                       (156, "S34", 1)]
    
    for i in data_Components:
        con.execute("INSERT INTO Components VALUES (?,?,?)", i)

con.commit()
```

```python
con.execute("UPDATE Products SET Cost=3.50 WHERE ProductName='Pink Kitten'")
con.execute("DELETE FROM Components WHERE CompID='G56'")
con.execute("INSERT INTO Products VALUES (?,?,?,?)", (190, "Fluffy puppy", 4.00, 8.00))
con.execute("INSERT INTO Components VALUES (?,?,?)", (190, "ST01", 20))
con.execute("INSERT INTO Components VALUES (?,?,?)", (190, "G34", 3))


con.commit()
```

```python
import pandas.io.sql as pds

query =pds.read_sql("SELECT a.ProductName, b.CompID, b.CompQty FROM Products AS a INNER JOIN Components AS b ON a.ProductID=b.ProductID",
                    con)

query
```
